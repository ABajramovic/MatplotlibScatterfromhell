## Total Fares by City Type

```python
# Calculate Type Percents
Total_Fares = pyber_df.groupby(["type"])["fare"].sum()
Total_Fares

# Build Pie Chart
Total_Fares.index

colors=["gold", "skyblue", "coral"]
explode= (0,0,0)

plt.pie(Total_Fares, labels =Total_Fares.index, 
        autopct="%1.2f%%", colors=colors,explode=explode, shadow=True, startangle=0)
# Save Figure

```

## Total Rides by City Type

```python
# Calculate Type Percents
Total_Fares = pyber_df.groupby(["type"])["ride_id"].count()
Total_Fares

# Build Pie Chart
Total_Fares.index

colors=["gold", "skyblue", "coral"]
explode= (0,0,0.5)

plt.pie(Total_Fares, labels =Total_Fares.index, 
        autopct="%1.2f%%", colors=colors,explode=explode, shadow=False, startangle=140)

# Save Figure


```

## Total Drivers by City Type

```python
# Calculate Type Percents
Total_Fares = pyber_df.groupby(["type"])["driver_count"].mean()
Total_Fares

# Build Pie Chart
Total_Fares.index

colors=["gold", "skyblue", "coral"]
explode= (0,0.5,0)

plt.pie(Total_Fares, labels =Total_Fares.index, 
        autopct="%1.2f%%", colors=colors,explode=explode, shadow=True, startangle=300)

# Save Figure
```
